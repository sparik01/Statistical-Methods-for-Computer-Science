```r
#x <- lm(formula = TotalCases ~ POPEST_2018_B_U18 + POPEST_2018_B_18TO64 + 
#    POPEST_2018_B_65PLUS + POPEST_2018_B_85PLUS + UNEMPLOYED + 
 #   VETPOP_2019_65TO84 + VETPOP_2019_85PLUS + VETPOP_2019_TOTAL + 
 #  HOUSING_UNITS + DENSITY_POP + DENSITY_HOUSE + PER_DIABETES + 
 #   OVERCROWDING + PER_RURAL + POP_RURAL + PER_UNINSURED + PER_AREA_WATER + 
 #   MED_INCOME_2018 + PER_COLLEGE_EDU + INCOME_RATIO + 
 # PER_UNEMPLOYED, data = data3)

x <- lm(formula = TotalCases ~  
    POPEST_2018_B_65PLUS + POPEST_2018_B_U18 + UNEMPLOYED +
    DENSITY_POP + DENSITY_HOUSE + POP_RURAL + HOUSING_UNITS + PER_DIABETES +
    MED_INCOME_2018 + PER_COLLEGE_EDU, data = data3)

summary(x)

par(mfrow = c(2,2))
plot(x)
```

```r
tidy(x)
```

```r
anova(x)
```

```r
#predict what happens to cases if the population of those under 18 goes down
predict(x, data.frame("POPEST_2018_B_65PLUS" = 1000, "POPEST_2018_B_U18" = 20816, "UNEMPLOYED" = 942,
                     "DENSITY_POP" = 91.8, "DENSITY_HOUSE" = 37.2, "POP_RURAL" = 22921, "HOUSING_UNITS" = 22135, "PER_DIABETES" = 0.111,
                      "MED_INCOME_2018" = 58786 , "PER_COLLEGE_EDU" = 0.62))
```

```r
log_x = lm(log1p(TotalCases) ~  
    log1p(POPEST_2018_B_65PLUS) + log1p(POPEST_2018_B_U18) + log1p(UNEMPLOYED) +
    log1p(DENSITY_POP) + log1p(DENSITY_HOUSE) + log1p(POP_RURAL) + log1p(HOUSING_UNITS) + log1p(PER_DIABETES) +
    log1p(MED_INCOME_2018) + log1p(PER_COLLEGE_EDU), data = data3)
summary(log_x)

par(mfrow = c(2,2))
plot(x)
```

```r
#Data Review and Data Prearation

#Data partition
set.seed(1255)

#70% in first data set and 30% in another - independent samples
samples <- sample(2, nrow(data3), 
              replace = TRUE, 
              prob = c(0.70, 0.30))

#nrow(data3) - 3136
training <- data3[samples==1,]
#nrow(training) - 2163
testing <- data3[samples==2,]
#nrow(testing) - 973

#summary of test and training data
cbind(summary(training$TotalCases), summary(testing$TotalCases))
```

```r
#model using training data set

xtraining <- lm(TotalCases ~  
    POPEST_2018_B_65PLUS + POPEST_2018_B_U18 + UNEMPLOYED +
    DENSITY_POP + DENSITY_HOUSE + POP_RURAL + HOUSING_UNITS + PER_DIABETES +
    MED_INCOME_2018 + PER_COLLEGE_EDU, data = training)

summary(xtraining)

par(mfrow = c(2,2))
plot(x)

```

```r
# making predictions
prediction <- predict(xtraining, testing)
head(prediction)
head(testing)
```

```r
#multicollinearity
#if VIF > 10 - presence of multicolinearity
#install.packages("faraway")
library(faraway)
vif(xtraining)
```

```r
#model after removing multicolinearity
xnew <- lm(TotalCases ~  
    POPEST_2018_B_U18 +
    DENSITY_POP + POP_RURAL + PER_DIABETES +
    MED_INCOME_2018 + PER_COLLEGE_EDU, data = training)

summary(xnew)

par(mfrow = c(2,2))
plot(xnew)
vif(xnew)
```

```r
prediction2 <- predict(xnew, testing)
head(prediction2)
head(testing)
```

```r
#percent error
per_error <- function(pred,actual) {
    abs((pred-actual)/actual)*100
}

#pred1
pred1 <- per_error(6451,8733)

#pred2
pred2 <- per_error(1025,1165)

#pred3
pred3 <- per_error(3874,6301)

#pred4
pred4 <- per_error(1259,422)

#pred5
pred5 <- per_error(711,919)

#pred6
pred6 <- per_error(894,711)

#vector of percent error
vec <- c(pred1, pred2, pred3, pred4, pred5, pred6)

mean(vec)

```
